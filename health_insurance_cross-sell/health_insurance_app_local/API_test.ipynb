## get_prediction

```python
#model
import pickle
model = pickle.load(open('models/xgbclassifier_model.pkl', 'rb'))
model
```

```python
#original_data (test_raw there)
d_test_raw = df_test.copy()
d_test_raw
```

```python
#test_data
d3
```

```python
#model prediction
pred = model.predict_proba(d3)
pred
```

```python
#join prediction into original data and sort
d_test_raw['score'] = pred[:, 1].tolist()
d_test_raw = d_test_raw.sort_values('score', ascending=False)
d_test_raw
```

```python
return_get_prediction = d_test_raw.to_json( orient= 'records', date_format = 'iso' )
return_get_prediction
```
